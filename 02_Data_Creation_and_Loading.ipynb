```python
# ==============================
# PROJECT: Placement Eligibility Streamlit Application
# FILE: 02_Data_Creation_Loading.ipynb
# PURPOSE: Create & Load Dataset, Handle Missing Values
# ==============================

# ------------------------------
# 1️⃣ Import Libraries
# ------------------------------
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

# ------------------------------
# 2️⃣ Dataset Creation
# ------------------------------
# Example: Placement Eligibility Dataset
data = {
    "Student_ID": [101, 102, 103, 104, 105],
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [22, 23, np.nan, 24, 22],                  # np.nan for missing value
    "Gender": ["F", "M", "M", "M", "F"],
    "Graduation_Score": [75, 82, 65, np.nan, 90],    # Missing value included
    "Backlogs": [0, 1, 2, 0, 0],
    "Internship": ["Yes", "No", "Yes", "No", "Yes"],
    "Placed": ["Yes", "No", "No", "Yes", "Yes"]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# ------------------------------
# 3️⃣ Save Dataset as CSV
# ------------------------------
csv_file = "placement_eligibility_dataset.csv"
df.to_csv(csv_file, index=False)
print(f"Dataset saved as {csv_file} ✅")

# ------------------------------
# 4️⃣ Load Dataset
# ------------------------------
df_loaded = pd.read_csv(csv_file)

# ------------------------------
# 5️⃣ Handle Missing Values
# ------------------------------
# Filling numeric missing values with mean
df_loaded['Age'].fillna(df_loaded['Age'].mean(), inplace=True)
df_loaded['Graduation_Score'].fillna(df_loaded['Graduation_Score'].mean(), inplace=True)

# Confirm changes
print("Missing values handled ✅")
print(df_loaded)

# ------------------------------
# 6️⃣ Verification
# ------------------------------
print("Dataset creation and loading completed successfully ✅")
```
