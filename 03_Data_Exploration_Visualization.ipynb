```python
# ==============================
# 03_Data_Exploration_Visualization.ipynb
# PROJECT: Placement Eligibility Streamlit Application
# PURPOSE: Explore dataset and visualize distributions and correlations
# ==============================

# ------------------------------
# 1️⃣ Import Libraries
# ------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

# ------------------------------
# 2️⃣ Load Dataset
# ------------------------------
csv_file = "placement_eligibility_dataset.csv"
df = pd.read_csv(csv_file)

# Display first 5 rows
df.head()
```

```python
# ------------------------------
# 3️⃣ Dataset Info
# ------------------------------
print("✅ Dataset Info:")
df.info()

print("\n✅ Check for Missing Values:")
print(df.isnull().sum())

# ------------------------------
# 4️⃣ Handle Missing Values
# ------------------------------
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Graduation_Score'].fillna(df['Graduation_Score'].mean(), inplace=True)

print("\n✅ Missing values after handling:")
print(df.isnull().sum())
```

```python
# ------------------------------
# 5️⃣ Summary Statistics
# ------------------------------
print("✅ Numeric Data Description:")
display(df.describe())

print("\n✅ Categorical Data Description:")
display(df.describe(include='object'))
```

```python
# ------------------------------
# 6️⃣ Histograms for Numeric Features
# ------------------------------
numeric_cols = ['Age', 'Graduation_Score', 'Backlogs']
df[numeric_cols].hist(figsize=(10,6), bins=10, color='skyblue', edgecolor='black')
plt.suptitle("Histogram of Numeric Features", fontsize=16)
plt.show()
```

```python
# ------------------------------
# 7️⃣ Countplots for Categorical Features
# ------------------------------
categorical_cols = ['Gender', 'Internship', 'Placed']

plt.figure(figsize=(15,4))
for i, col in enumerate(categorical_cols):
    plt.subplot(1, 3, i+1)
    sns.countplot(x=col, data=df, palette='Set2')
    plt.title(f'Countplot of {col}')
plt.tight_layout()
plt.show()
```

```python
# ------------------------------
# 8️⃣ Correlation Heatmap (Numeric Features Only)
# ------------------------------

# Select only numeric columns
numeric_df = df.select_dtypes(include=np.number)

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='Blues', fmt=".2f")
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

```python
# ------------------------------
# 9️⃣ Boxplots
# ------------------------------
plt.figure(figsize=(10,4))
sns.boxplot(data=df[numeric_cols], palette='Set3')
plt.title("Boxplot of Numeric Features")
plt.show()
```

```python
# ------------------------------
# 10 Boxplots
# ------------------------------


# Placement vs Graduation Score
plt.figure(figsize=(6,4))
sns.boxplot(x='Placed', y='Graduation_Score', data=df, palette='pastel')
plt.title("Graduation Score by Placement Status")
plt.show()

# Placement vs Age
plt.figure(figsize=(6,4))
sns.boxplot(x='Placed', y='Age', data=df, palette='pastel')
plt.title("Age by Placement Status")
plt.show()
```

```python
# ------------------------------
# 11 Boxplots
# ------------------------------


plt.figure(figsize=(6,4))
sns.countplot(x='Gender', hue='Placed', data=df, palette='Set1')
plt.title("Placement Status by Gender")
plt.show()
```

```python
# ------------------------------
# 12 Boxplots
# ------------------------------

plt.figure(figsize=(6,4))
sns.countplot(x='Internship', hue='Placed', data=df, palette='Set1')
plt.title("Placement Status by Internship")
plt.show()
```
