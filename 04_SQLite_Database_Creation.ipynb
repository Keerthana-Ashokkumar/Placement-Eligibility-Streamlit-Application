```python
# ------------------------------
# 04_SQLite_Database_Creation.ipynb
# PROJECT: Placement Eligibility Streamlit Application
# PURPOSE: Create SQLite DB, Tables, and Populate with Faker Data
# ------------------------------

import sqlite3
from faker import Faker
import random
import pandas as pd
import datetime

# Suppress warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
# ------------------------------
# 1️⃣ Create SQLite Connection
# ------------------------------
conn = sqlite3.connect("placement_eligibility.db")
cursor = conn.cursor()
print("✅ SQLite Database Connected")
```

```python
# ------------------------------
# 2️⃣ Create Students Table
# ------------------------------
cursor.execute("""
CREATE TABLE IF NOT EXISTS Students (
    student_id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    age INTEGER,
    gender TEXT,
    email TEXT UNIQUE,
    phone TEXT,
    enrollment_year INTEGER,
    course_batch TEXT,
    city TEXT,
    graduation_year INTEGER
)
""")
print("✅ Students Table Created")

# ------------------------------
# 3️⃣ Create Programming Table
# ------------------------------
cursor.execute("""
CREATE TABLE IF NOT EXISTS Programming (
    programming_id INTEGER PRIMARY KEY,
    student_id INTEGER,
    language TEXT,
    problems_solved INTEGER,
    assessments_completed INTEGER,
    mini_projects INTEGER,
    certifications_earned INTEGER,
    latest_project_score REAL,
    FOREIGN KEY(student_id) REFERENCES Students(student_id)
)
""")
print("✅ Programming Table Created")

# ------------------------------
# 4️⃣ Create Soft Skills Table
# ------------------------------
cursor.execute("""
CREATE TABLE IF NOT EXISTS SoftSkills (
    soft_skill_id INTEGER PRIMARY KEY,
    student_id INTEGER,
    communication INTEGER,
    teamwork INTEGER,
    presentation INTEGER,
    leadership INTEGER,
    critical_thinking INTEGER,
    interpersonal_skills INTEGER,
    FOREIGN KEY(student_id) REFERENCES Students(student_id)
)
""")
print("✅ Soft Skills Table Created")

# ------------------------------
# 5️⃣ Create Placements Table
# ------------------------------
cursor.execute("""
CREATE TABLE IF NOT EXISTS Placements (
    placement_id INTEGER PRIMARY KEY,
    student_id INTEGER,
    mock_interview_score INTEGER,
    internships_completed INTEGER,
    placement_status TEXT,
    company_name TEXT,
    placement_package REAL,
    interview_rounds_cleared INTEGER,
    placement_date TEXT,
    FOREIGN KEY(student_id) REFERENCES Students(student_id)
)
""")
print("✅ Placements Table Created")

# Commit table creation
conn.commit()
```

```python
# ------------------------------
# 6️⃣ Initialize Faker
# ------------------------------
fake = Faker()
Faker.seed(42)
random.seed(42)

# Number of students
num_students = 20  
```

```python
# Generate Students Data

students_data = []
for i in range(1, num_students+1):
    name = fake.name()
    age = random.randint(20, 28)
    gender = random.choice(["Male", "Female", "Other"])
    email = fake.unique.email()
    phone = fake.phone_number()
    enrollment_year = random.randint(2019, 2023)
    course_batch = f"DS Batch {random.randint(1,5)}"
    city = fake.city()
    graduation_year = enrollment_year + 4
    students_data.append((i, name, age, gender, email, phone, enrollment_year, course_batch, city, graduation_year))

df_students = pd.DataFrame(students_data, columns=[
    "student_id","name","age","gender","email","phone",
    "enrollment_year","course_batch","city","graduation_year"
])

df_students.head()
```

```python
# Generate programming Data

programming_data = []

languages = ["Python", "SQL", "R", "Java"]

for student in df_students['student_id']:
    language = random.choice(languages)
    problems_solved = random.randint(10, 100)
    assessments_completed = random.randint(1, 10)
    mini_projects = random.randint(0, 5)
    certifications_earned = random.randint(0, 3)
    latest_project_score = round(random.uniform(60, 100), 2)
    programming_data.append((None, student, language, problems_solved, assessments_completed, mini_projects, certifications_earned, latest_project_score))

df_programming = pd.DataFrame(programming_data, columns=[
    "programming_id","student_id","language","problems_solved",
    "assessments_completed","mini_projects","certifications_earned","latest_project_score"
])

df_programming.head()
```

```python
# Generate Soft Skills Data

softskills_data = []

for student in df_students['student_id']:
    communication = random.randint(50, 100)
    teamwork = random.randint(50, 100)
    presentation = random.randint(50, 100)
    leadership = random.randint(50, 100)
    critical_thinking = random.randint(50, 100)
    interpersonal_skills = random.randint(50, 100)

    softskills_data.append((
        None, student, communication, teamwork, presentation,
        leadership, critical_thinking, interpersonal_skills
    ))

df_softskills = pd.DataFrame(softskills_data, columns=[
    "soft_skill_id","student_id","communication","teamwork",
    "presentation","leadership","critical_thinking","interpersonal_skills"
])

df_softskills.head()
```

```python
#  Generate Placement Data

placement_data = []

for student in df_students['student_id']:
    mock_interview_score = random.randint(50, 100)
    internships_completed = random.randint(0, 3)
    placement_status = random.choice(["Ready", "Not Ready", "Placed"])

    if placement_status == "Placed":
        company_name = fake.company()
        placement_package = round(random.uniform(3.0, 15.0), 2)  # Lakhs per annum example
        interview_rounds_cleared = random.randint(1, 5)
        placement_date = fake.date_between(start_date="-1y", end_date="today")
    else:
        company_name = None
        placement_package = None
        interview_rounds_cleared = 0
        placement_date = None

    placement_data.append((
        None, student, mock_interview_score, internships_completed,
        placement_status, company_name, placement_package,
        interview_rounds_cleared, placement_date
    ))

df_placements = pd.DataFrame(placement_data, columns=[
    "placement_id","student_id","mock_interview_score","internships_completed",
    "placement_status","company_name","placement_package",
    "interview_rounds_cleared","placement_date"
])

df_placements.head()
```

```python
df_students.to_sql("Students", conn, if_exists="replace", index=False)
df_programming.to_sql("Programming", conn, if_exists="replace", index=False)
df_softskills.to_sql("SoftSkills", conn, if_exists="replace", index=False)
df_placements.to_sql("Placements", conn, if_exists="replace", index=False)

conn.commit()
print("🔄 Tables replaced and data saved!")
```

```python
# ------------------------------
# Step 6: SQL Queries & Insights
# ------------------------------

# Import necessary libraries
import sqlite3
import pandas as pd
import warnings

# Suppress warnings for clean output
warnings.filterwarnings("ignore")

# Establish SQLite connection
conn = sqlite3.connect("placement_eligibility.db")
cursor = conn.cursor()
print("✅ Connected to SQLite Database")
```

```python
def run_query(query, conn=conn):
    """Execute SQL query and return a pandas DataFrame"""
    df = pd.read_sql_query(query, conn)
    print(f"✅ Query executed: {query.split()[0]} ...")  
    return df
```

```python
 # 3️⃣ Queries with Insights
```

```python
# Query 1: Average Programming Performance per Batch

query1 = """
SELECT s.course_batch,
       ROUND(AVG(p.latest_project_score),2) AS avg_project_score,
       ROUND(AVG(p.problems_solved),2) AS avg_problems_solved
FROM Students s
LEFT JOIN Programming p ON s.student_id = p.student_id
GROUP BY s.course_batch
ORDER BY avg_project_score DESC
"""
avg_prog_batch = run_query(query1)
avg_prog_batch
```

```python
# Query 2: Top 5 Students by Programming Score

query2 = """
SELECT s.name, s.course_batch, p.latest_project_score, p.problems_solved
FROM Students s
JOIN Programming p ON s.student_id = p.student_id
ORDER BY p.latest_project_score DESC
LIMIT 5
"""
top_programmers = run_query(query2)
top_programmers
```

```python
# Query 3: Students Ready for Placement

query3 = """
SELECT s.name, s.course_batch, p.latest_project_score, pl.placement_status
FROM Students s
JOIN Programming p ON s.student_id = p.student_id
JOIN Placements pl ON s.student_id = pl.student_id
WHERE pl.placement_status = 'Ready'
ORDER BY p.latest_project_score DESC
"""
ready_students = run_query(query3)
ready_students
```

```python
# Query 4: Top Soft Skills Students

query4 = """
SELECT s.name,
       (ss.communication + ss.teamwork + ss.presentation + ss.leadership + ss.critical_thinking + ss.interpersonal_skills)/6 AS avg_softskills
FROM Students s
JOIN SoftSkills ss ON s.student_id = ss.student_id
ORDER BY avg_softskills DESC
LIMIT 5
"""
top_softskills = run_query(query4)
top_softskills
```

```python
# Query 5: Average Soft Skills per Batch

query5 = """
SELECT s.course_batch,
       ROUND(AVG(ss.communication),2) AS avg_communication,
       ROUND(AVG(ss.teamwork),2) AS avg_teamwork,
       ROUND(AVG(ss.presentation),2) AS avg_presentation,
       ROUND(AVG(ss.leadership),2) AS avg_leadership,
       ROUND(AVG(ss.critical_thinking),2) AS avg_critical_thinking,
       ROUND(AVG(ss.interpersonal_skills),2) AS avg_interpersonal_skills
FROM Students s
JOIN SoftSkills ss ON s.student_id = ss.student_id
GROUP BY s.course_batch
ORDER BY avg_communication DESC
"""
softskills_batch = run_query(query5)
softskills_batch
```

```python
# Query 6: Placement Status Distribution

query6 = """
SELECT placement_status, COUNT(*) AS student_count
FROM Placements
GROUP BY placement_status
"""
placement_counts = run_query(query6)
placement_counts
```

```python
# Query 7: Average Placement Package for Placed Students

query7 = """
SELECT ROUND(AVG(placement_package),2) AS avg_package
FROM Placements
WHERE placement_status = 'Placed'
"""
avg_package = run_query(query7)
avg_package
```

```python
# Query 8: High Programming Score but Not Placed

query8 = """
SELECT s.name, p.latest_project_score, pl.mock_interview_score, pl.placement_status
FROM Students s
JOIN Programming p ON s.student_id = p.student_id
JOIN Placements pl ON s.student_id = pl.student_id
WHERE p.latest_project_score > 85 AND pl.placement_status != 'Placed'
ORDER BY p.latest_project_score DESC
"""
high_prog_low_place = run_query(query8)
high_prog_low_place
```

```python
# Query 9: Internship Count vs Placement Status

query9 = """
SELECT internships_completed, placement_status, COUNT(*) AS count
FROM Placements
GROUP BY internships_completed, placement_status
ORDER BY internships_completed DESC
"""
internship_vs_placement = run_query(query9)
internship_vs_placement
```

```python
# Query 10: Students Needing Soft Skills Improvement

query10 = """
SELECT s.name, ss.communication, ss.teamwork, ss.presentation, pl.placement_status
FROM Students s
JOIN SoftSkills ss ON s.student_id = ss.student_id
JOIN Placements pl ON s.student_id = pl.student_id
WHERE ss.communication < 60 OR ss.teamwork < 60 OR ss.presentation < 60
ORDER BY ss.communication ASC
"""
students_improve = run_query(query10)
students_improve
```
